```python
# Data
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

# Task 1: Mengambil daftar semua perusahaan dan mengurutkannya secara terbalik
company_names = sorted([{'name': company['name']} for company in company_detail_list], key=lambda x: x['name'], reverse=True)
print("Task 1:", company_names)

# Task 2: Mencetak domain setiap perusahaan
for company in company_detail_list:
    print(f"{company['name']}: {company['domain']} ({company['country']})")

# Task 3: Mengelompokkan karyawan berdasarkan domain perusahaan
employees_by_domain = {}
domain_map = {company['name']: company['domain'] for company in company_detail_list}

for employee in employee_detail_list:
    domain = domain_map.get(employee['company'])
    if domain:
        employees_by_domain.setdefault(domain, []).append(employee['full_name'])

print("Task 3:", employees_by_domain)

# Task 4: Fungsi untuk mendapatkan karyawan dan negara perusahaan
def get_employees_with_company_country():
    employees_with_country = []
    country_map = {company['name']: company['country'] for company in company_detail_list}

    for employee in employee_detail_list:
        country = country_map.get(employee['company'])
        employees_with_country.append({
            "full_name": employee['full_name'],
            "company": employee['company'],
            "country": country
        })

    return employees_with_country

print("Task 4:", get_employees_with_company_country())

# Task 5: Fungsi untuk menghitung jumlah karyawan berdasarkan kewarganegaraan di setiap perusahaan
def count_employees_by_nationality():
    nationality_count_by_company = []
    for company in company_detail_list:
        nationality_count = {}

        for employee in employee_detail_list:
            if employee['company'] == company['name']:
                nationality = employee['nationality']
                nationality_count[nationality] = nationality_count.get(nationality, 0) + 1

        nationality_count_by_company.append({
            "company": company['name'],
            "employee_nationality": nationality_count
        })

    return nationality_count_by_company

print("Task 5:", count_employees_by_nationality())
```
